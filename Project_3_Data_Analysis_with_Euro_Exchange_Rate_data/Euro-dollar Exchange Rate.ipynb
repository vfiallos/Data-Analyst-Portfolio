# 1. Introduction

The goal of this analysis is to tell a story on the following:
* We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline.
* We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for 2006 and 2009 for comparison.
* We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)).

#### Data Set Information
The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the US dollar is 1.5, you get 1.5 US dollars if you pay 1.0 euro (one euro has more value than one US dollar at this exchange rate).

# 5. Data Cleaning

### Renaming columns and changing types

```python
# rename [US dollar] & period\unit 
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
```

```python
# change time column to a datetime data type
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

# sort values by time in ascending order
exchange_rates.sort_values('Time', inplace=True)

# reset the index and drop initial index
exchange_rates.reset_index(drop=True, inplace=True)
```

### Extracting attributes for the Euro-dollar rate

```python
#isolating time and us_dollar column
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

Notice the '-' has a count of 62. This doesn't seem right for a dollar value.

### Dropping rows with ' - '

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

### US dollar to float type

```python
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

# 6. Data Analysis

### Euro-dollar rate progression

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
```

If we look at the line's shape, we see many small wiggles — rather than seeing a smooth line. The wiggles, however, have meaning: they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

Depending on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the **rolling mean** (also known as the moving average).

### Rolling Mean

```python
#the window is 30 days and adding a new column to euro_to_dollar
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

We have added a new column to our euro to dollar dataset that displays the rolling mean with a window of 30 days. We want to minimize the harsh lines by using a rolling mean to better represent our data.

The first 30 rows will have NaN values as the window period is 30 days.

```python
#making a grid chart that displays different rolling windows
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

## Coronoa Virus

```python
pre_covid = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2016
                   ) & (euro_to_dollar['Time'].dt.year <= 2020)]
covid = euro_to_dollar.copy(
                   )[euro_to_dollar['Time'].dt.year == 2020]      
```

```python
pre_covid.describe()
```

```python
# Adding the FiveThirtyEight style
style.use('fivethirtyeight')

# Adding the plot
fig,ax = plt.subplots(figsize=(10,4))
ax.plot(pre_covid['Time'],
        pre_covid['rolling_mean'],
        linewidth=1, color='#A6D785')
# Highlighting the 2020 period
ax.plot(covid['Time'],
        covid['rolling_mean'],
        linewidth=3, color='#e23d28')

# Adding seperate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 16750.0
for year in ['2016', '2017', '2018', '2019', '2020','2021']:
    ax.text(x, 1.03, year, alpha=0.5, fontsize=11)
    x += 365

y = 1.047
for rate in ['1','1.06','1.12','1.18']:
    ax.text(16600.0, y, rate, alpha=0.5, fontsize=11)
    y += 0.05
    
# Highlihting the peak of the crisis
ax.axvspan(18400, 18628,ymin=0.03,
           alpha=0.3, color='grey')
    
# Adding a title and a subtitle    
ax.text(16600, 1.28, "Euro-USD rate peaked at 1.23 during Corona Virus Outbreak",
        weight='bold')
ax.text(16600, 1.26, 'Euro-USD exchange rates between 2016 and 2020',
        size=12)

# Adding a signature
ax.text(16600.0, 1, '©DATAQUEST' + ' '*140 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

# Add some transparency to the grid
ax.grid(alpha=0.4)
```

**The pandemic started around March 2020, and we see the euro-dollar rate head upwards to a peak around that time.**

## The Financial Crisis

```python
financial_crisis.describe()
```

```python
### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Adding the plot
fig,ax = plt.subplots(figsize=(10,4))
ax.plot(financial_crisis['Time'],
        financial_crisis['rolling_mean'],
        linewidth=1, color='#A6D785')
### Highlighting the 2007-2008 period
ax.plot(financial_crisis_7_8['Time'],
        financial_crisis_7_8['rolling_mean'],
        linewidth=3, color='#e23d28')

# Adding seperate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 13110.0
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha=0.5, fontsize=11)
    x += 365

y = 1.195
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(13010.0, y, rate, alpha=0.5, fontsize=11)
    y += 0.1
    
#### TRIAL AND ERROR    
### Highlihting the peak of the crisis
ax.axvspan(13937.25, 14139, ymin=0.09,
           alpha=0.3, color='grey')

## Adding a title and a subtitle    
ax.text(13010.0, 1.64, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        weight='bold')
ax.text(13010.0, 1.61, 'Euro-USD exchange rates between 2006 and 2010',
        size=12)

### Adding a signature
ax.text(13010.0, 1.07, '©DATAQUEST' + ' '*140 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

### Add some transparency to the grid
ax.grid(alpha=0.4)

plt.show()
```

## Three Presidents Viz

```python
bush_obama_trump = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy(
       )[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

Below, you'll notice we used matplotlib's functional approach to build the graphs. We use this approach because it offers more flexibility in arranging the subplots:

* We first build three of the graphs on a 2-by-3 grid (this grid should have six subplots, but we only build three; the bottom row remains empty).
* We then build only the bottom graph of a 2-by-1 grid (this grid should have two subplots; the top row remains empty).
* The two grids are merged, and we end up with three graphs on the top row and one graph on the bottom row.

```python
#style.available
```

```python
### Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Adding the subplots
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

### Changes to all the subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)
    ax.grid(alpha=0.5) 
    
### Ax1: Bush
## use color #65765e when done
ax1.plot(bush['Time'], bush['rolling_mean'],
        color='#65765e')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'],
                   alpha=0.3)
ax1.text(12418, 1.92, 'BUSH', fontsize=18, weight='bold',
        color='#65765e')
ax1.text(12095, 1.8, '(2001-2009)', weight='bold',
        alpha=0.3)

### Ax2: Obama
#### use color #dd966b when done
ax2.plot(obama['Time'], obama['rolling_mean'],
        color='#dd966b')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'],
                   alpha=0.3)
ax2.text(15150, 1.92, 'OBAMA', fontsize=18, weight='bold',
        color='#dd966b')
ax2.text(15000, 1.8, '(2009-2017)', weight='bold',
         alpha=0.3)

### Ax3: Trump
### use color #6667ab when done
ax3.plot(trump['Time'], trump['rolling_mean'],
        color='#6667ab')
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '',
                     '2020', '', '2021'],
                   alpha=0.3)
ax3.text(17630, 1.92, 'TRUMP', fontsize=18, weight='bold',
        color='#6667ab')
ax3.text(17550, 1.8, '(2017-2021)', weight='bold',
         alpha=0.3)

### Ax4: Bush-Obama-Trump
ax4.plot(bush['Time'], bush['rolling_mean'],
        color='#65765e')
ax4.plot(obama['Time'], obama['rolling_mean'],
        color='#dd966b')
ax4.plot(trump['Time'], trump['rolling_mean'],
        color='#6667ab')
ax4.grid(alpha=0.5)
ax4.set_xticks([])

### Adding a title and a subtitle
ax1.text(10800, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(10800, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=16)

### Adding a signature
ax4.text(10700, 0.65, '©DATAQUEST' + ' '*103 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()
```

**The bottom graph is most telling with all line plots combined, we can easily see how the rates changed along with the Presidents of the United States. The euro-dollar rate went up during the Bush term, stayed around 1.4 and lower during Obama, and stayed below 1.2 during Trump.**
